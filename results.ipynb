```python
datasets = ['coauthor', 'squirrel', 'reddit2', 'arxiv', 'patents']
models = ['vanilla']

metric_val = "val_f1"
metric_test = "test_f1" 

# for each combination of dataset and model get the best validation auroc and print the test auroc
for dataset in datasets:
	for model in models:

		# ENSEMBLE
		# best_val = df[df["Sweep"] == f'{dataset}_{model}'].sort_values("auroc_AU_classic", ascending=False).head(1)
		# print(f"Test AU AUROC for {dataset} with classical {model}: {best_val['auroc_AU_classic'].values[0]}")

		# best_val = df[df["Sweep"] == f'{dataset}_{model}'].sort_values("auroc_AU_credal", ascending=False).head(1)
		# print(f"Test AU AUROC for {dataset} with credal {model}: {best_val['auroc_AU_credal'].values[0]}")

		# best_val = df[df["Sweep"] == f'{dataset}_{model}'].sort_values("auroc_EU_classic", ascending=False).head(1)
		# print(f"Test EU AUROC for {dataset} with classical {model}: {best_val['auroc_EU_classic'].values[0]}")

		# best_val = df[df["Sweep"] == f'{dataset}_{model}'].sort_values("auroc_EU_credal", ascending=False).head(1)
		# print(f"Test EU AUROC for {dataset} with credal {model}: {best_val['auroc_EU_credal'].values[0]}")


		# CREDAL MODELS
		
		# best_val = df[df["Sweep"] == f'{dataset}_{model}'].sort_values("val_auroc_EU", ascending=False).head(1)
		# print(f"Test AUROC (EU) for {dataset} with {model}: {best_val['test_auroc_EU'].values[0]}")

		# VANILLA



		# Consider only the first 30 created runs for this dataset/model
		subset = df[df["Sweep"] == f'{dataset}_{model}'].sort_values("Created").head(30)
		best_val = subset.dropna(subset=[metric_val]).sort_values(metric_val, ascending=False).head(1)
		if not best_val.empty:
			print(f"{metric_test} for {dataset} with {model} (first 30 runs): {best_val[metric_test].values[0]}")
		else:
			print(f"No valid runs for {dataset} with {model} in the first 30 created.")

		print()

```
